```python
data_saniti=pd.read_csv('/content/Sanitation1.csv')
data_df_working=pd.DataFrame(data_saniti)
data_df_working
```

```python
data_df_working['Dry_Resourse_Collection_Center_(Yes/No)'] = data_df_working['Dry_Resourse_Collection_Center_(Yes/No)'].replace(np.nan,'YES')
```

```python
data_df_working['Dry_Resourse_Collection_Center_(Yes/No)'] = data_df_working['Dry_Resourse_Collection_Center_(Yes/No)'].replace(0,'YES')
```

```python
data_df_working['Dry_Resourse_Collection_Center_(Yes/No)'].unique()
```

```python
data_df_working[12:17]
```

```python
data_df_working['toilets']=(data_df_working['Public_Toilets']+data_df_working['Community_Toilets']+data_df_working['She_Toilets'])
```

```python
column_names = ['Public_Toilets', 'Community_Toilets', 'She_Toilets']
data_df_working['toilets']= data_df_working[column_names].sum(axis=1)
```

```python
data_df_working
```

```python
data_df_working['no_of_wards']=data_df_working['ODF_Declared_Wards']+data_df_working['ODF_Non_Declared_Wards']
```

```python
data_df_working
```

```python
data_df_working['%gar_gen_lf']=(data_df_working['Garbage_lifted/day_(Mts)']/data_df_working['Garbage_generation/day_(Mts)'])*100
```

```python
data_df_working
```

```python

corr_mat=data_df_working.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr_mat,annot=True,cmap='Greys')
```

```python
data_df_working['responsibility']=(data_df_working['ODF_Declared_Wards']/data_df_working['no_of_wards'])*100
data_df_working
```
